```python
import numpy as np 

def calculate_person_angle(a,b):
    if a is None:
        print("Problem in Chest")
        angle = 0
        
    elif b is None:
        print("Problem in Neck")
        angle = 0
    
    else:        
        a = np.array(a) # First
        b = np.array(b) # Neck
    
        radians = np.arctan2(a[1]-b[1], 0) - np.arctan2(a[1]-b[1], a[0]-b[0])
        angle = np.abs(radians*180.0/np.pi)
                
        if (b[0] - a[0]) < 0:
            angle = angle * -1
        
    return angle 

def calculate_left_arm_angle(a,b):
    if a is None:
        print("Problem in LWrist")
        angle = 0
        
    elif b is None:
        print("Problem in RShoulder")
        angle = 0
    else:
        a = np.array(a) # Left Wrist
        b = np.array(b) # Left Shoulder
    
        # 90 graus - (Angulo entre a reta dos pontos e o eixo Y)
        radians = np.pi/2 - np.arctan2(a[1]-b[1], a[0]-b[0])
        angle = np.abs(radians*180.0/np.pi)
    
        if angle >180.0:
            print ('ERROR!')
        
    return angle 

def calculate_right_arm_angle(a,b):
    if a is None:
        print("Problem in RWrist")
        angle = 0
        
    elif b is None:
        print("Problem in LShoulder")
        angle = 0
    
    else:
        a = np.array(a) # Right Wrist
        b = np.array(b) # Right Shoulder
    
        # 90 graus - (Angulo entre a reta dos pontos e o eixo Y)
        radians = np.pi/2 - np.arctan2(a[1]-b[1], a[0]-b[0])
        angle = np.abs(radians*180.0/np.pi)
    
        if angle >180.0:
            angle = 360-angle
        
    return angle 
```
