```python
import import_ipynb
import lateral_raise as LR
import barbell_curl as BP
import print_result as PR

BODY_PARTS = { "Head": 0, "Neck": 1, "RShoulder": 2, "RElbow": 3, "RWrist": 4,
               "LShoulder": 5, "LElbow": 6, "LWrist": 7, "DHip": 8, "RKnee": 9,
               "DAnkle": 10, "LHip": 11, "LKnee": 12, "LAnkle": 13, "Chest": 14 }

LATERAL_RAISE_WARNING = { "None": 0, "Right Arm": 1, "Left Arm": 2, "Both Arms": 3 }

BARBELL_CURL_WARNING = { "None": 0, "Hiperextension": 1, "Right Arm": 2, "Left Arm": 3, "Both Arms": 4 }

def confiability_upper(confs,prob_min_lateral_raise):
    Num_Points = 9
    
    Sum_Confiability = (confs[BODY_PARTS["Head"]] + confs[BODY_PARTS["Neck"]]+ confs[BODY_PARTS["RShoulder"]] 
                    + confs[BODY_PARTS["RElbow"]] + confs[BODY_PARTS["RWrist"]]+ confs[BODY_PARTS["LShoulder"]] 
                    + confs[BODY_PARTS["LElbow"]] + confs[BODY_PARTS["LWrist"]] + confs[BODY_PARTS["Chest"]])
    
    Confiability = ((Sum_Confiability)/Num_Points)
    
    print("Confaibilidade = " + str(Confiability))
    
    return False if (Confiability <= prob_min_lateral_raise) else True
    
    
        
def inspec_lateral_raise(confs,prob_min_lateral_raise,frame, points, angle_max_arms, print_angles):
    if confiability_upper(confs,prob_min_lateral_raise):
        warning = LATERAL_RAISE_WARNING["None"]
    
        bodyAngle = LR.calculate_person_angle(points[BODY_PARTS["Chest"]],points[BODY_PARTS["Neck"]])
        
        angleRight = LR.calculate_right_arm_angle(points[BODY_PARTS["RWrist"]],points[BODY_PARTS["LShoulder"]])
        
        if angleRight > (angle_max_arms+bodyAngle) : 
            warning = LATERAL_RAISE_WARNING["Right Arm"]
            
        angleLeft = LR.calculate_left_arm_angle(points[BODY_PARTS["LWrist"]],points[BODY_PARTS["RShoulder"]])
        
        if angleLeft > (angle_max_arms-bodyAngle) : 
            if warning == LATERAL_RAISE_WARNING["Right Arm"]:
                warning = LATERAL_RAISE_WARNING["Both Arms"]
            else:
                warning = LATERAL_RAISE_WARNING["Left Arm"]
        
        frame = PR.print_lateral_raise_result(frame, points, warning)
    
        if(print_angles):
            frame = PR.print_body_angle(frame, bodyAngle, points)
            frame = PR.print_arms_angle(frame, angleRight, angleLeft, points)
            
def inspec_barbell_curl(confs,prob_min_barbell_curl,frame, points, angle_accepted_body, angle_accepted_arms, print_angles):
    if confiability_upper(confs,prob_min_barbell_curl):
        warning = BARBELL_CURL_WARNING["None"]
        
        bodyAngle = BP.calculate_person_angle(points[BODY_PARTS["RShoulder"]],points[BODY_PARTS["LShoulder"]],
                                              points[BODY_PARTS["Neck"]],points[BODY_PARTS["Chest"]])
        
        if bodyAngle > angle_accepted_body : 
            warning = BARBELL_CURL_WARNING["Hiperextension"]
            
        angleRight = BP.calculate_right_arm_angle(points[BODY_PARTS["RWrist"]],points[BODY_PARTS["RShoulder"]],
                                                  points[BODY_PARTS["LShoulder"]])
        
        if angleRight > (bodyAngle+angle_accepted_arms) : 
            warning = BARBELL_CURL_WARNING["Right Arm"]
        
        angleLeft = BP.calculate_left_arm_angle(points[BODY_PARTS["RShoulder"]],points[BODY_PARTS["LShoulder"]],
                                                points[BODY_PARTS["LWrist"]])
        
        if angleLeft > (bodyAngle+angle_accepted_arms) : 
            if warning == BARBELL_CURL_WARNING["Right Arm"]:
                warning = BARBELL_CURL_WARNING["Both Arms"]
            else:
                warning = BARBELL_CURL_WARNING["Left Arm"]
            
        frame = PR.print_barbell_curl_result(frame, points, warning)
              
        if(print_angles):
            frame = PR.print_body_angle(frame, bodyAngle, points)
            frame = PR.print_arms_angle(frame, angleRight, angleLeft, points)
    
```
