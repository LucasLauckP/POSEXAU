```python
import numpy as np 

def calculate_person_angle(a,b,c,d):
    a = np.array(a) # Right Shoulder
    b = np.array(b) # Left Shoulder
    c = np.array(c) # Neck
    d = np.array(d) # Chest
    # intersection point calculation
    m1 = (b[1] - a[1])/(b[0] - a[0])

    if ( d[0] - c[0] ) != 0:
        m2 = (d[1] - c[1])/(d[0] - c[0])
        xi = ( m1 * a[0] - m2 * c[0] - a[1] ) / (m1 - m2)
        
    else:
        xi = c[0]

    yi = m1 * (xi - a[0]) + a[1]
    
    radians = np.arctan2(d[1]-yi, d[0]-xi) - np.arctan2(a[1]-yi, a[0]-xi)
    angle = np.abs(radians*180.0/np.pi)
    
    if angle >180.0:
        angle = 360-angle
        
    return angle 

def calculate_left_arm_angle(a,b,c):
    if a is None:
        print("Problem in LWrist")
        angle = 0
        
    elif b is None:
        print("Problem in LShoulder")
        angle = 0
        
    elif c is None:
        print("Problem in RShoulder")
        angle = 0
        
    a = np.array(a) # Left Wrist
    b = np.array(b) # Left Shoulder
    c = np.array(c) # Right Shoulder
    
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)
    
    if angle >180.0:
        angle = 360-angle
        
    return angle

def calculate_right_arm_angle(a,b,c):
    if a is None:
        print("Problem in LShoulder")
        angle = 0
        
    elif b is None:
        print("Problem in RShoulder")
        angle = 0
        
    elif c is None:
        print("Problem in RWrist")
        angle = 0
        
    a = np.array(a) # Left Shoulder
    b = np.array(b) # Right Shoulder
    c = np.array(c) # Right Wrist
    
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)
    
    if angle >180.0:
        angle = 360-angle
        
    return angle
```
